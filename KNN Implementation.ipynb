```python
import pandas as pd
import numpy as np
df = pd.read_csv('./archive/MachineLearningCSV/MachineLearningCVE/_Friday-WorkingHours-Afternoon-DDos.pcap_ISCX - Copy.csv')
df = df.fillna(0)
df = df.replace([np.inf, -np.inf], 9999999999999)
```

```python
X = df.iloc[:, :-1].values # all columns except the last one are for training dataset
y  = df[' Label']
y
```

```python
# spliting our dataset to 80% training data and 20% testing data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
# scaling our training and testing dataset
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# fitting data to our model
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

score_sum = 0
n = 100
for i in range(0, n): 
    y_pred = classifier.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    score_sum = score_sum + score

avg_score = score_sum/100
avg_score
```
