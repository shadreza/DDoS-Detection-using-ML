```python
import pandas as pd
import numpy as np
```

```python
# df = pd.read_csv('./archive/MachineLearningCSV/MachineLearningCVE/_Friday-WorkingHours-Afternoon-DDos.pcap_ISCX - Copy.csv')
df = pd.read_csv('./archive/MachineLearningCSV/MachineLearningCVE/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
# df = pd.read_csv('diabetes.csv')
df
```

```python
df = df.replace([np.inf, -np.inf], 9999999999999)
```

```python
df = df.fillna(0)
```

```python
df.isna().sum().sum()
```

```python
X = df.drop(columns=[' Label']) 
#input set created by splitting the data, denote with capital letter - convention
X
```

```python
y  = df[' Label'] #creating o/p dataset - another way than dropping
y
```

```python
oneRow = X.loc[[20]]
oneRow
```

```python
twoRow = X.loc[[212320]]
twoRow
```

```python
from sklearn.tree import DecisionTreeClassifier 
#scikit learning library for ML with python

model = DecisionTreeClassifier() 
model.fit(X, y) #that's all we need, just passing the ip set and op set to the model

predictions = model.predict(twoRow)
predictions 
```

```python
#calculating accuracy. Divide entire ds to 80/20 - train + test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.2)

model.fit(X_train, y_train) #that's all we need, just passing the ip set and op set to the model

score_sum = 0
n = 100
for i in range(0, n): 
    predictions = model.predict(X_test) 
    score = accuracy_score(y_test, predictions)
    score_sum = score_sum + score

avg_score = score_sum/100
avg_score
```
